Explication des modèles

Description des modèles et choix

Régression Logistique : C'est un modèle qui calcule la probabilité qu'une description appartienne à une catégorie en attribuant un "poids" à chaque mot. Je l'ai choisi car c'est la base incontournable pour la classification de texte : c'est simple, rapide et on comprend tout de suite quels mots influencent le résultat.

Random Forest : Ce modèle utilise une multitude d'arbres de décision qui votent pour la meilleure catégorie. Je l'ai choisi pour voir si des combinaisons de mots (plus complexes que de simples poids) permettaient de mieux détecter les contenus matures.

SVM : Le SVM cherche à tracer la frontière la plus nette possible entre les deux groupes. C'est un modèle réputé pour être très efficace quand on a beaucoup de variables (ici, des milliers de mots différents), ce qui est le cas avec notre vectorisation TF-IDF.

Hypothèses

L'hypothèse principale est que le vocabulaire utilisé dans le résumé Netflix reflète directement la classification d'âge. On suppose que les résumés pour adultes utilisent un ton ou des thèmes (drogue, violence, humour noir) absents des résumés tout public.

Variables les plus importantes

En observant les coefficients des modèles, on voit clairement ce qui fait pencher la balance :

Pour le contenu Mature : Les mots comme "drug", "sex", "murder", "horror" et "violent".

Pour le contenu Général : Les mots comme "kids", "animated", "save", "adventures", "magical" et "competition".

Le modèle associe donc la maturité à la violence, aux thèmes sociaux adultes, tandis que le contenu général est lié à  l'aventure et à la famille.

Analyse d'erreurs

Malgré une précision correcte (environ 65%), le modèle fait des erreurs systématiques :

Les thrillers "propres" : Un film d'action avec la CIA ou un avocat (exemple : The Lincoln Lawyer) est souvent classé en "Général" par le modèle car le résumé utilise des termes neutres (justice, équipe, sécurité), alors que le film lui-même contient de la violence classée R.

Le vocabulaire trompeur : Un documentaire ou un film familial qui parle de "terreur" ou de "menace" de façon métaphorique peut être classé à tort en "Mature".

Manque de contexte : Le modèle ne comprend pas l'ironie ou le second degré. Si une description de film pour enfant dit "ils doivent affronter leurs peurs", le mot "peurs" peut induire le modèle en erreur.

La Régression Logistique s'en sort le mieux car elle est moins sujette au "sur-apprentissage" (overfitting) sur ce petit volume de texte par rapport au Random Forest.
